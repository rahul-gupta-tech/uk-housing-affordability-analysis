```python
import pandas as pd
df = pd.read_csv("../data/ONS_regional_fulltime_salary_data.csv") ## ../ <-- goes back one folder, then into the data folder to obtain the regional full-time data
print(df.head(10)) #shows the first 5 rows. e.g., df.head(10) shows the first 10 rows
```

```python
#In order to join the data with the Median House Prices by County, cleaning of County Names is required.

df['Description'] = (
    df['Description']
    .str.upper() #makes local authority name upper case
    .str.strip() #removes leading or trailing spaces
    .str.replace(' UA', '', regex=False) #removes "_UA" from end of local authority name
    .str.replace(' MET COUNTY', '', regex=False) #removes "Met County" from end of local authority name
)
```

```python
print(df.head())
df['Median'] = pd.to_numeric(
    df['Median'], #converts median to number
    errors='coerce') #found error in median salary for two local authorities - in spreadhseet, these were labelled x. Changed to NaN values
df[df['Median'].isna()]
```

```python
df = df.dropna(subset=['Median']) #dropping NaN rows for easier joining
df[df['Median'].isna()]
```

```python
df = df.rename(columns={
    'Description': 'local_authority',
    'Code': 'ons_code',
    'Median': 'median_salary'
})
```
